```python
#!/usr/bin/env python3

# Ensure you install the required packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Initialize an unauthenticated client for public datasets
client = Socrata("data.wa.gov", None)

# Initialize variables
limit = 50000
offset = 0
all_results = []

while True:
    # Fetch a chunk of data
    results = client.get("rpr4-cgyd", limit=limit, offset=offset)
    
    # If no more data is returned, break the loop
    if not results:
        break
    
    # Append this chunk to the list
    all_results.extend(results)
    
    # Increment the offset to fetch the next chunk
    offset += limit

# Convert the full list of results to a DataFrame
results_df = pd.DataFrame.from_records(all_results)

# Save the DataFrame to a CSV file
results_df.to_csv("washington_ev_data_full.csv", index=False)

print("All data saved to washington_ev_data_full.csv")
```

# Exploratory Analysis 


Our analysis aims to forecast the number of new electric vehicles being added to the roads in each county over time. As a result, we are focused solely on new vehicle purchases, excluding used vehicle transactions. However, any vehicle initially purchased as new in Washington will still be represented in our dataset, even if it later becomes "used."

To determine the number of electric vehicles on the road over time, we calculate the cumulative sum of monthly "Original Title" transactions for each county and organize this data into a dataframe.

```python
vehicle_count = {}
for region in df['county'].unique():
    monthly_data = df[
        (df['county'] == region) &
        (df['transaction_type'] == 'Original Title')
    ].resample('ME').size().cumsum()
    
    vehicle_count[region] = monthly_data
```

```python
# Filling null values created by resampling with 0's
cumulative_df = pd.DataFrame(vehicle_count)
cumulative_df.fillna(0, inplace=True)
cumulative_df.head()
```

## Most Popular Electric Vehicle Models by County

To assist an electric vehicle charging company in making informed decisions, understanding the most popular EV models purchased within each county can provide valuable insights. Key factors that can be influenced by this data include:

### Adapter Requirements: 
Determining the types of adapters to install, such as Tesla, CCS, and CHAdeMO, to support the charging needs of diverse vehicle models.
### Station Types:
Assessing which levels of charging stations to build (level 1, level 2, or fast-charging level 3) to meet the demands of the most commonly used vehicles in the area.
### Charger Spacing and Layout: 
Deciding on the layout and spacing of chargers based on the typical vehicle size in each county, ensuring sufficient space and convenience for users.
By analyzing model popularity trends across counties, the charging infrastructure can be optimized to provide tailored services that meet local needs effectively.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_top_models_by_county(data, num_models=10, counties=None):
    # Ensure counties list is provided
    if counties is None:
        print("Counties list is missing.")
        return

    # Set up a 2x5 grid for 10 counties
    fig, axes = plt.subplots(figsize=(20, 16), nrows=5, ncols=2, sharex=False, sharey=False)
    axes = axes.flatten()  # Flatten the axes array for easy indexing
    
    # Iterate through the counties and create subplots
    for idx, county_name in enumerate(counties):
        # Filter data for each county and transaction type
        county_data = data[(data['county'] == county_name) & 
                           (data['transaction_type'] == 'Original Title')]
        
        if county_data.empty:
            print(f"No data for county: {county_name}")
            axes[idx].text(0.5, 0.5, "No Data", ha='center', va='center', fontsize=12, color="red")
            axes[idx].axis('off')
            continue
        
        # Count the top vehicle models
        model_counts = pd.DataFrame(county_data['model'].value_counts()).reset_index()
        model_counts.columns = ['vehicle_model', 'purchase_count']
        
        # Plot the barplot on the grid
        sns.barplot(x='purchase_count', y='vehicle_model', data=model_counts.head(num_models),
                    ax=axes[idx], palette='coolwarm')
        
        # Set titles and labels
        axes[idx].set_title(f'Top {num_models} Models in {county_name} County (2010-2024)', fontsize=12)
        axes[idx].set_xlabel('Purchase Count', fontsize=10)
        axes[idx].set_ylabel('Model', fontsize=10)

    # Hide any unused subplots if `counties` has fewer than 10 items
    for j in range(idx + 1, len(axes)):
        axes[j].axis('off')
    
    # Adjust layout and add a main title
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)
    plt.suptitle('Top Electric Vehicle Models by County (2010-2024)', fontsize=18, color='darkred')
    
    plt.savefig('./vehicle_model_trends_by_county.png', facecolor='white')

    plt.show()  # Display the plot

# Example usage with placeholders for `data` and `counties`
# Replace `df` and `top_10` with your actual DataFrame and list of counties
# plot_top_models_by_county(data=df, num_models=10, counties=top_10)
# Assuming you have a DataFrame named `df` and a list of counties named `top_10`
# Example call:
plot_top_models_by_county(data=df, num_models=10, counties=top_10)
```

```python




import pandas as pd

# Load the zip code to county mapping CSV into a DataFrame
zip_to_county_df = pd.read_csv('./ZIP-COUNTY-FIPS_2017-06.csv')  # Replace with your actual file path

# Filter for zip codes in Washington state (WA)
zip_to_county_wa_df = zip_to_county_df[zip_to_county_df['STATE'] == 'WA']

# Ensure zip codes are treated as strings for matching
zip_to_county_wa_df['ZIP'] = zip_to_county_wa_df['ZIP'].astype(str)

# Convert the filtered DataFrame to a dictionary for fast lookup
zip_to_county_dict = pd.Series(zip_to_county_wa_df['COUNTYNAME'].values, index=zip_to_county_wa_df['ZIP']).to_dict()

# Function to get county by zip code with only WA zip codes, excluding "Unknown" counties
def get_county_by_zip_wa(zip_codes):
    """
    Given a list of zip codes, returns a dictionary mapping WA zip codes to their respective county names,
    excluding entries with "Unknown" counties.
    
    Parameters:
        zip_codes (list): List of zip codes to search for counties.
    
    Returns:
        dict: A dictionary with WA zip codes as keys and county names as values.
    """
    zip_county_map = {}
    for zip_code in zip_codes:
        # Convert zip_code to string to match dictionary keys
        zip_code_str = str(zip_code)
        # Get county name and exclude "Unknown" entries
        county_name = zip_to_county_dict.get(zip_code_str)
        if county_name and county_name != "Unknown":  # Only add if county is not "Unknown"
            zip_county_map[zip_code_str] = county_name
    
    return zip_county_map

# Sample usage
# Convert zip codes to strings and filter WA zip codes in your data if necessary
zipcodes_to_search = df_charging_stations[df_charging_stations['state'] == 'WA']['zip'].astype(str).unique()
zip_county_map = get_county_by_zip_wa(zipcodes_to_search)

# Display the zip-to-county mapping without "Unknown" counties
zip_county_map


```

```python
import pandas as pd
import folium
import random
import matplotlib.colors as mcolors

# Load the county charger count data from the CSV file
file_path = 'county_charger_counts_2024-10-29.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Initialize a Folium map focused on Washington State
m = folium.Map(location=[47.5, -120.5], zoom_start=7, tiles="CartoDB positron")

# Generate a unique color for each county
colors = list(mcolors.TABLEAU_COLORS.values())
random.shuffle(colors)  # Shuffle to randomize color selection
color_map = {county: colors[i % len(colors)] for i, county in enumerate(df['County'])}

# Smaller radius for individual bubbles
bubble_radius = 0.5  # Small, fixed size for each charger bubble

# Function to add random offset with increased spread for greater spacing
def add_random_offset(latitude, longitude, spread=0.5):
    lat_offset = random.uniform(-spread, spread)
    lon_offset = random.uniform(-spread, spread)
    return latitude + lat_offset, longitude + lon_offset

# Plot each charger as a separate small bubble
for _, row in df.iterrows():
    county_color = color_map[row['County']]  # Get a unique color for each county
    
    # Set spread range based on charger count for wider dispersion in large counts
    spread_range = 0.5 if row['Count'] < 500 else 0.8  # Larger spread for high counts
    
    for _ in range(int(row['Count'])):
        # Randomize position with an even larger spread within the county area
        lat, lon = add_random_offset(row['Latitude'], row['Longitude'], spread=spread_range)
        
        # Add each charger as a small bubble with reduced size
        folium.CircleMarker(
            location=[lat, lon],
            radius=bubble_radius,
            color=county_color,
            fill=True,
            fill_color=county_color,
            fill_opacity=0.7
        ).add_to(m)

# Save map to HTML file or display directly in Jupyter Notebook
m.save("washington_ev_chargers_distributed_map.html")
m  # If running in Jupyter, this will display the map
```
